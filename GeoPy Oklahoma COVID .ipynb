```python
### install packages ###
import pandas as pd
import geopy
from geopy.distance import geodesic 
import numpy as np
from pandas import DataFrame
```

```python
### View dataframe ###
cities_df = pd.read_excel('XXXXXXX.xlsx')
cities_df.head()
```

```python
cities_df.shape
```

```python
## This creates a function to refer to the lattitude of a city ###
def lat(City):
    ic=(cities_df[cities_df["city"]==City].index.values)
    lat2 = cities_df.loc[ic,'lat'].values[0]
    return lat2
```

```python
### This creates a function to refer to the longitude of a city ###
def lngt(City):
    ic=(cities_df[cities_df["city"]==City].index.values)
    lng2 = cities_df.loc[ic,'lng'].values[0]
    return lng2
```

```python
ok = lngt("Oklahoma City")
print(ok)
```

```python
### This creates a function to calculate the distance of each city from eachother geospatially ###
def distancefrom2coodf(City):
    #(a,b)=getcoord(City)
    a=lat(City)
    b=lngt(City)
    rowsfunc = []
    i = 0
    while i<739: 
        lati = cities_df.loc[i, 'lat']
        long = cities_df.loc[i, 'lng']
        coords_1 = (lati, long)
        coords_2 = (a, b)
        distance_okc = geodesic(coords_1, coords_2).miles
        rowsfunc.append(distance_okc)
        i+=1
    return rowsfunc
```

```python
cities_df['tulsa_distance'] = distancefrom2coodf("Tulsa")
cities_df['okc_distance'] = distancefrom2coodf("Oklahoma City")
cities_df['Norman_distance'] = distancefrom2coodf("Norman")
cities_df['BrokenArrow_distance'] = distancefrom2coodf("Broken Arrow")
cities_df['Edmond_distance'] = distancefrom2coodf("Edmond")
cities_df['Lawton_distance'] = distancefrom2coodf("Lawton")
cities_df['Moore_distance'] = distancefrom2coodf("Moore")
cities_df['Midwest_distance'] = distancefrom2coodf("Midwest City")
cities_df['Stw_distance'] = distancefrom2coodf("Stillwater")
cities_df['enid_distance'] = distancefrom2coodf("Enid")
```

```python
### This creates distance from closest city with at least 100,000 population ###
j=0
closemetro =[]
while j<739: 
    closestmetro=min(cities_df.loc[j,'tulsa_distance'],cities_df.loc[j,'okc_distance'])
    closemetro.append(closestmetro)
    j+=1
cities_df['ht_metro']=closemetro
```

```python
### This creates distance from closest city with at least 40,000 population. ###
l=0
closemetro2 =[]
while l<739: 
    closestmetro2=min(cities_df.loc[l,'tulsa_distance'],cities_df.loc[l,'okc_distance'],cities_df.loc[l,'Norman_distance'],cities_df.loc[l,'BrokenArrow_distance'],cities_df.loc[l,'Edmond_distance'], cities_df.loc[l,'Lawton_distance'],cities_df.loc[l,'Moore_distance'],cities_df.loc[l,'Midwest_distance']
                      ,cities_df.loc[l,'Stw_distance']
                      ,cities_df.loc[l,'enid_distance'] )
    closemetro2.append(closestmetro2)
    l+=1
cities_df['ft_metro']=closemetro2
```

```python
m=0
closemetro1 =[]
while m<739: 
    mminvalue=min(cities_df.loc[m,'tulsa_distance'],
                                             cities_df.loc[m,'okc_distance'],
                                             cities_df.loc[m,'Norman_distance'],
                                             cities_df.loc[m,'BrokenArrow_distance'],
                                             cities_df.loc[m,'Edmond_distance'],
                                             cities_df.loc[m,'Lawton_distance'],
                                             cities_df.loc[m,'Moore_distance'],
                                             cities_df.loc[m,'Midwest_distance'],
                                             cities_df.loc[m,'Stw_distance'],
                                             cities_df.loc[m,'enid_distance'])
    
    if(cities_df.loc[m,'tulsa_distance']==mminvalue):
        kbc='Tulsa'
    elif(cities_df.loc[m,'okc_distance']==mminvalue):
            kbc='Oklahoma City'
    elif(cities_df.loc[m,'Norman_distance']==mminvalue):
            kbc='Norman'
    elif(cities_df.loc[m,'BrokenArrow_distance']==mminvalue):
            kbc='Broken Arrow'
    elif(cities_df.loc[m,'Edmond_distance']==mminvalue):
            kbc='Edmond'
    elif(cities_df.loc[m,'Lawton_distance']==mminvalue):
            kbc='Lawton'
    elif(cities_df.loc[m,'Moore_distance']==mminvalue):
            kbc='Moore'
    elif(cities_df.loc[m,'Midwest_distance']==mminvalue):
            kbc='Midwest City'
    elif(cities_df.loc[m,'Stw_distance']==mminvalue):
            kbc='Stillwater'
    elif(cities_df.loc[m,'enid_distance']==mminvalue):
            kbc='Enid'

    closemetro1.append(kbc)
    m+=1
cities_df['ft_metro_name']=closemetro1
```
